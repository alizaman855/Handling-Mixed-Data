```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('titanic2.csv')
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
fig=df['number'].value_counts().plot.bar()
fig.set_title("Passengers travelling with")
```

```python
#Extracting numerical Part
df['number_numerical']=pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python
df['number_categorical']=np.where(df['number_numerical'].isnull(),df['number'],np.nan)
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['cabin_num']=df['Cabin'].str.extract('(\d+)') #Capture Numerical Part
df['cabin_cat']=df['Cabin'].str[0]     #Capture the first letter
```

```python
df.head()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar',color='red')
```

```python
#Extracting the last bit of Ticket
df['ticket_num']=df['Ticket'].apply(lambda s:s.split()[-1])
df['ticket_num']=pd.to_numeric(df['ticket_num'],errors='coerce',downcast='integer')
df.head()
```

```python
#Extracting the first part of Ticket
df['ticket_cat']=df['Ticket'].apply(lambda s:s.split()[0])
df['ticket_cat']=np.where(df['ticket_cat'].str.isdigit(),np.nan,df['ticket_cat'])
df.head()
```
